```python
# Ä°NTEGER, FLOAT VE STRINGLER
print(9)
print(9.2)
print(9 + 9)
print(9 * 9)

print("HELLO AI ERA")

print(type(9))      # integer
print(type(9.2))    # float
print(type("HELLO AI ERA"))  # string

# STRINGLERE YAXINDAN BAXAQ

print("")
print('')

print(123)
print(type(123))

print("123")
print(type("123"))

print("a" + "b")
print("a" " b")
print("a" + "-b")

# print("a" - "b")  # Error: Stringler cixila bilmez

print("a " * 3)
# print("a" / 3)  # Error: Stringler bolune bilmez

# STRING METODLARI - len()

gel_yaz = "gelecegi_yazanlar"

print(len(gel_yaz))
print(len("gelecegi_yazanlar"))

# STRING METODLARI - upper() & lower()

gel_yaz = "gelecegi_yazanlar"

print(gel_yaz.upper())
print(gel_yaz.lower())

print(gel_yaz.islower())

B = gel_yaz.upper()

print(B.islower())
print(B.isupper())

# STRING METODLARI - replace()

gel_yaz = "gelecegi_yazanlar"

print(gel_yaz.replace("e", "a"))
print(gel_yaz.replace("a", "i"))

# STRING METODLARI - strip()

gel_yaz = " gelecegi_yazanlar "
print(gel_yaz.strip())

gel_yaz = "*gelecegi_yazanlar*"
print(gel_yaz.strip("*"))

gel_yaz = "lgelecegi_yazanlarl"
print(gel_yaz.strip("l"))

# METODLARA UMUMI BAXIS

gel_yaz = "gelecegi_yazanlar"

print(dir(gel_yaz))

print(gel_yaz.capitalize())
print(gel_yaz.title())

# ALTSTRINGLER

gel_yaz = "gelecegi_yazanlar"

print(gel_yaz[1])

# print(gel_yaz[20])  # Xeta: Indeks serhedi kecir

print(gel_yaz[0:3])

# VARIABLES (DEYISENLER)

a = 9
b = "TURAL KOSMOSA GETDI"
c = a * 2

print(a / c)
print(a * c)
print(a * 5)

print(type(100))
print(type(100.2))
print(type(1 + 2j))
```
